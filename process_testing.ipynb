```python
import pathlib
import numpy as np
import pandas as pd
```

```python
data_path = r"C:\I24 Motion Project\YB Outlier Detection\csv_data\TM_chunk0_ratio0_GT.csv"

df = pd.read_csv(data_path, skiprows=0, index_col=False)
df = df.reindex(columns=['Frame #', 'ID', 'x'])
```

```python
df = df.loc[df['ID'] == 100]
df = df.sort_values(by='Frame #').reset_index(drop=True)

pd.set_option('display.max_rows', df.shape[0]+1)
display(df)
```

```python
x_train = []
batch_size = 100

start_row, end_row = 0, int(df.size / 3)

while(end_row - start_row >= batch_size):
    df_batch = df.iloc[start_row : start_row + batch_size]
    batch_ar = np.array(df_batch['x'])
    x_train.append(batch_ar)
    start_row += batch_size

x_train = np.array(x_train)
display(x_train)
# num_of_batch = int(x_train.size / batch_size)
```
